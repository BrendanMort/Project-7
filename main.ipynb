```python
import csv
import datetime

# Modified from https://automatetheboringstuff.com/chapter14/
def process_csv(filename):
    example_file = open(filename, encoding="utf-8")
    example_reader = csv.reader(example_file)
    example_data = list(example_reader)
    example_file.close()
    return example_data

vaccination_data = process_csv("vaccinations.csv")
header = vaccination_data[0]
covid_data = vaccination_data[1:]


def cell(row_idx,col_name):
    col_idx = header.index(col_name)
    val = covid_data[row_idx][col_idx]
    if val == "":
        return None
    return val
    
header

    
```

```python
#q1
unique_dates = []
for idx in range(len(covid_data)):
    if cell(idx,"date") not in unique_dates:
        unique_dates.append(cell(idx,"date"))
        
unique_dates
```

```python
#q2
countries = []
for idx in range(len(covid_data)):
    if cell(idx,"country") not in countries:
        countries.append(cell(idx,"country"))
sorted(countries)
```

```python
#q3
populations = {}
for idx in range(len(covid_data)):
    country = cell(idx,"country")
    population = cell(idx,"population")
    if country not in populations:
        populations[country] = None 
    if population != None:
        populations[country] = int(population)
populations
```

```python
#q4
vax_data_12 = []
for idx in range(len(covid_data)):
    if (cell(idx, "date") == "10/12/2021") and (cell(idx,"daily_vaccinations") != None) and (cell(idx,"country") not in vax_data_12):
        vax_data_12.append(cell(idx,"country"))
len(vax_data_12)       
```

```python
def daily_vaccinations_on(date):
    '''return a dictionary mapping each country to the number of daily vaccinations on the given date; if data is missing, the value for that country is None'''
    daily_vax_dict = {}
    for i in range(len(covid_data)):
        country_i = cell(i,"country")
        date_i = cell(i,"date")
        daily_vax_i = cell(i,"daily_vaccinations")
        if date_i == date: 
            if daily_vax_i != None:
                daily_vax_dict[country_i] = int(daily_vax_i)
            else:
                daily_vax_dict[country_i] = daily_vax_i
    return daily_vax_dict
```

```python
#q6
weekly_vax = {}
for country in daily_vaccinations_on("10/13/2021").keys():
    weekly_vax[country] = 0
for date in unique_dates:
    for i in daily_vaccinations_on(date):
        if daily_vaccinations_on(date)[i] != None:
            weekly_vax[i] += daily_vaccinations_on(date)[i]
for i in weekly_vax:
    if weekly_vax[i] == 0:
        weekly_vax[i] = None
weekly_vax
```

```python
#q7
max_vax = 0
max_country = None
for i in weekly_vax:
    if (weekly_vax[i] != None):
        if (weekly_vax[i] > max_vax):
            max_vax = weekly_vax[i]
            max_country = i
max_country        
```

```python
#q8
vax_by_date = {}

for date in unique_dates:
    vax_by_date[date] = 0

for date in unique_dates:
    total = 0
    for i in daily_vaccinations_on(date):
        if daily_vaccinations_on(date)[i] != None:
            total += daily_vaccinations_on(date)[i]
    vax_by_date[date] = total
        
vax_by_date
```

```python
#q9
last_vax_date = {}
for i in range(len(covid_data)):
    last_vax_date[cell(i,"country")] = None
for i in range(len(covid_data)):
    country_i = cell(i,"country")
    
    if cell(i,"people_fully_vaccinated") != None:
        last_vax_date[country_i] = cell(i,"date")
        
last_vax_date

```

```python
def get_number_of_days(start_date, end_date):
    day1 = datetime.datetime.strptime(start_date, '%m/%d/%Y') 
    day2 = datetime.datetime.strptime(end_date, '%m/%d/%Y')
    delta = day2 - day1
    return delta.days

def fully_vaccinated_by(date):
    fully_vax_dict = {}
    for i in range(len(covid_data)):
        country_i = cell(i,"country")
        date_i = cell(i,"date")
        fully_vax = cell(i,"people_fully_vaccinated")
        if country_i not in fully_vax_dict:
            fully_vax_dict[country_i] = None
        if fully_vax == None:
            continue
        if get_number_of_days(date,date_i) <= 0:
            fully_vax_dict[country_i] = int(fully_vax)
    return fully_vax_dict   
        

```

```python
#q11
max_country_under_100 = None
max_vax = 0
countries_under_100 = []
for i in range(len(covid_data)):
    if cell(i,"population") != None:
        if int(cell(i,"population")) < 100000000 and cell(i,"country") not in countries_under_100:
            countries_under_100.append(cell(i,"country"))
for i in countries_under_100:
    if fully_vaccinated_by("10/14/2021")[i] != None:
        if fully_vaccinated_by("10/14/2021")[i] > max_vax:
            max_country_under_100 = i
            max_vax = fully_vaccinated_by("10/14/2021")[i]
max_country_under_100


max_country_under_100
```

```python
def people_vaccinated_by(date):
    people_vax_dict = {}
    for i in range(len(covid_data)):
        country_i = cell(i,"country")
        date_i = cell(i,"date")
        people_vax = cell(i,"people_vaccinated")
        if country_i not in people_vax_dict:
            people_vax_dict[country_i] = None
        if people_vax == None:
            continue
        if get_number_of_days(date,date_i) <= 0:
            people_vax_dict[country_i] = int(people_vax)
    return people_vax_dict   

def total_vaccinations_by(date):
    total_vax_dict = {}
    for i in range(len(covid_data)):
        country_i = cell(i,"country")
        date_i = cell(i,"date")
        total_vax = cell(i,"total_vaccinations")
        if country_i not in total_vax_dict:
            total_vax_dict[country_i] = None
        if total_vax == None:
            continue
        if get_number_of_days(date,date_i) <= 0:
            total_vax_dict[country_i] = int(total_vax)
    return total_vax_dict   

vaccination_stats = {}

for date in unique_dates: 
    daily_vax_dict = daily_vaccinations_on(date)  # a dict mapping each country to number of daily_vaccinations on date
    total_vax_dict = total_vaccinations_by(date) # a dict mapping each country to the most recent number of total_vaccinations on date
    people_vax_dict = people_vaccinated_by(date) # a dict mapping each country to the most recent number of people_vaccinated on date
    fully_vax_dict = fully_vaccinated_by(date) # a dict mapping each country to most recent number of people_fully_vaccinated on date
    for country in countries:
        if country not in vaccination_stats:
            vaccination_stats[country] = {}
        vaccination_stats[country][date] = {}
        vaccination_stats[country][date]['country'] = country
        vaccination_stats[country][date]['date'] = date
        vaccination_stats[country][date]['daily_vaccinations'] = daily_vax_dict[country]
        vaccination_stats[country][date]['total_vaccinations'] = total_vax_dict[country]
        vaccination_stats[country][date]["people_vaccinated"] = people_vax_dict[country]
        vaccination_stats[country][date]['people_fully_vaccinated'] = fully_vax_dict[country]
        vaccination_stats[country][date]["population"] = populations[country]
        
```

```python
#q17
ratio_vax = {}
for country in vaccination_stats:
    people_vax = vaccination_stats[country]["10/16/2021"]["people_vaccinated"]
    fully_vax = vaccination_stats[country]["10/16/2021"]["people_fully_vaccinated"]
    if country not in ratio_vax:
        ratio_vax[country] = None
    if fully_vax != None and people_vax != None:
        ratio_vax[country] = fully_vax/people_vax
ratio_vax
```

```python
#q18
min_ratio = 999
min_ratio_country = None
for country in ratio_vax:
    if ratio_vax[country] == None:
        continue
    if ratio_vax[country] < min_ratio:
        min_ratio = ratio_vax[country]
        min_ratio_country = country
min_ratio_country
```

```python
#q19
max_per_capita = 0
max_capita_country = None
for country in vaccination_stats:
    if vaccination_stats[country]["10/14/2021"]["total_vaccinations"] == None or vaccination_stats[country]["10/14/2021"]["population"] == None:
        continue
    if vaccination_stats[country]["10/14/2021"]["total_vaccinations"]/vaccination_stats[country]["10/14/2021"]["population"] > max_per_capita:
        max_per_capita = vaccination_stats[country]["10/14/2021"]["total_vaccinations"]/vaccination_stats[country]["10/14/2021"]["population"]
        max_capita_country = country
max_capita_country       
```
